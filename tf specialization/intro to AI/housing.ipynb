```python
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
housing = pd.read_csv("housing.csv\\housing.csv")
```

```python
housing
```

```python
housing = housing.dropna()
```

```python
op = housing["ocean_proximity"]
op_dict = {i:v for v, i in enumerate(op.unique())}
op_arr = np.array([op_dict[i] for i in op])
housing["ocean_proximity"] = op_arr
```

```python
housing
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
labels = housing["median_house_value"].to_numpy()
data = housing[["latitude", "longitude", "housing_median_age", "total_rooms", "total_bedrooms", "population", "households", "median_income", "ocean_proximity"]]
```

```python
data
```

```python
labels
```

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled.shape
```

```python
BATCH_SIZE = 128
batched_data = np.empty((0, 32, 9))
batched_labels = np.empty((0, 32))
i=0
for i in range(0, len(data)//32):
    batched_data = np.append(batched_data, data_scaled[i:i+32].reshape(1, 32, 9), axis=0)
    batched_labels = np.append(batched_labels, labels[i:i+32].reshape(1, 32), axis=0)
    i += 32
```

```python
batched_label_scaled = batched_labels / labels.max()
```

```python
batched_label_scaled, batched_label_scaled.shape
```

```python
638 * 0.9
```

```python
model = keras.Sequential([
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(128, kernel_regularizer="L1"), 
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(32), 
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(8), 
    keras.layers.Dense(1)
])
model.compile(optimizer=keras.optimizers.Adam(learning_rate=5e-4), loss=keras.losses.mse)
```

```python
model.fit(batched_data[:510], batched_label_scaled[:510], epochs=250, 
         validation_data=(batched_data[510:574], batched_label_scaled[510:574]), 
         callbacks=[keras.callbacks.EarlyStopping(monitor="val_loss", patience=10)])
```

```python
model.evaluate(batched_data[574:], batched_label_scaled[574:])
```

```python
preds = model.predict(batched_data[574:])
```
